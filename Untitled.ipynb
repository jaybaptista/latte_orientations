```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_hdf("data_obliquityAngles_z_m12i_res7100_NEW.h5")
```

```python
fig, ax = plt.subplots(3, dpi=200)
ax[0].plot(df.snapshot, df["disk"], "-o")
ax[1].plot(df.snapshot, df["5.disk"], "-o")
ax[2].plot(df.snapshot, df["virial"], "-o")
plt.show()
```

```python
fig, ax = plt.subplots(dpi=200)
ax.plot(df.snapshot, df["disk"], c="r", alpha=0.25)
ax.plot(df.snapshot, df["5.disk"], c="r", alpha=0.5)
ax.plot(df.snapshot, df["virial"], c="r", alpha=1)
plt.show()
```
