```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.family':'serif'})
```

```python
m12i_data = pd.read_hdf('Alignment_peri/data_lmcAlignment_peri_m12i.h5')
m12f_data = pd.read_hdf('Alignment_peri/data_lmcAlignment_peri_m12f.h5')
m12w_data = pd.read_hdf('Alignment_peri/data_lmcAlignment_peri_m12w.h5')
m12m_data = pd.read_hdf('Alignment_peri/data_lmcAlignment_peri_m12m.h5')
```

```python
m12i_data
```

```python
fig, ax = plt.subplots(1, dpi = 200, figsize = (6, 3.5), sharey = 'row')
r = np.arange(10, 310, step = 10)

ax_f = ax_i = ax_w = ax_m = ax

ax_f.plot(r, m12f_data['angle'],
         ms=2, c='blue', marker=10, label='m12f')

ax_i.plot(r, m12i_data['angle'],
         ms=2, c='red', marker='o', label='m12i', ls='dashed')

ax_m.plot(r, m12m_data['angle'],
         ms=2, c='green', marker='*', label='m12m', ls='dotted')

ax_w.plot(r, m12w_data['angle'],
         ms=5, c='orange', marker='+', label='m12w', ls='dashdot')

ax_f.set_xlabel('r [kpc]')
ax_f.set_ylabel('Alignment [deg]')

ax.legend(loc = 'upper left')

plt.show()
```
