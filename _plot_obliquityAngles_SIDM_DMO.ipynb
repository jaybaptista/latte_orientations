```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.family':'serif'})
```

```python
m12i_data = pd.read_hdf('_data_obliquityAngles_z0_m12i.h5')
m12f_data = pd.read_hdf('_data_obliquityAngles_z0_m12f.h5')
m12m_data = pd.read_hdf('_data_obliquityAngles_z0_m12m.h5')
m12i_data_DMO = pd.read_hdf('_data_obliquityAngles_z0_DMO_m12i.h5')
m12f_data_DMO = pd.read_hdf('_data_obliquityAngles_z0_DMO_m12f.h5')
m12m_data_DMO = pd.read_hdf('_data_obliquityAngles_z0_DMO_m12m.h5')
m12i_data_SIDM = pd.read_hdf('_data_obliquityAngles_z0_SIDM_m12i.h5')
m12f_data_SIDM = pd.read_hdf('_data_obliquityAngles_z0_SIDM_m12f.h5')
m12m_data_SIDM = pd.read_hdf('_data_obliquityAngles_z0_SIDM_m12m.h5')
```

```python
fig, ax = plt.subplots(3, dpi = 200, figsize = (8, 4), sharex=True)

r = np.arange(2, 402, step = 2)

ax_f, ax_i, ax_m = ax


ax_f.plot(r, m12f_data['angle'],
         ms=2, c='k', alpha=.5,  label='CDM+Stars', lw=3)

ax_f.plot(r, m12f_data_DMO['angle'],
         ms=2, c='blue',  label='DMO', lw=3)

ax_f.plot(r, m12f_data_SIDM['angle'],
         ms=2, c='dodgerblue',  label='SIDM+Stars', lw=3)

ax_i.plot(r, m12i_data['angle'],
         ms=2, c='k', alpha=.5,  label='m12i', lw=3)

ax_i.plot(r, m12i_data_DMO['angle'],
         ms=2, c='blue',  label='m12i', lw=3)

ax_i.plot(r, m12i_data_SIDM['angle'],
         ms=2, c='dodgerblue',  label='m12i', lw=3)

ax_m.plot(r, m12m_data['angle'],
         ms=2, c='k', alpha=.5,  label='m12m', lw=3)

ax_m.plot(r, m12m_data_DMO['angle'],
         ms=2, c='blue',  label='m12m', lw=3)

ax_m.plot(r, m12m_data_SIDM['angle'],
         ms=2, c='dodgerblue',  label='m12m', lw=3)

ax_f.legend(loc = 'upper left', prop={'size': 11}, bbox_to_anchor=(1., 1))

ax_f.axhline(20, c='k', alpha=.25)
ax_i.axhline(20, c='k', alpha=.25)
ax_m.axhline(20, c='k', alpha=.25)

ax_f.set_yticklabels([0, 20])
ax_i.set_yticklabels([0, 20])
ax_m.set_yticklabels([0, 20])

for ax_k in ax:
    ax_k.set_ylim(0,40)
    ax_k.axvline(60, alpha=.25, c='k', ls='--')
    ax_k.axvline(200, alpha=.25, c='k', ls='--')
    ax_k.axvline(400, alpha=.25, c='k', ls='--')


ax_f.text(65, 38, 'Gaia', rotation=90, size=11, va='top')
ax_f.text(205, 38, 'AS', rotation=90, size=11, va='top')
ax_f.text(405, 38, 'LSST', rotation=90, size=11, va='top')

ax_m.set_xlabel('r [kpc]', size=11)

fig.text(-.075, .5, 'm12f', transform=ax_f.transAxes, rotation=90, ha='center', va='center', size = 11, fontweight='bold')
fig.text(-.075, .5, 'm12i', transform=ax_i.transAxes, rotation=90, ha='center', va='center', size = 11, fontweight='bold')
fig.text(-.125, .5, '$\\theta$ [deg]', transform=ax_i.transAxes, rotation=90, ha='center', va='center', size = 11)
fig.text(-.075, .5, 'm12m', transform=ax_m.transAxes, rotation=90, ha='center', va='center', size = 11, fontweight='bold')

fig.tight_layout()
fig.subplots_adjust(hspace=0)
plt.savefig('obliquityAngles_z0_SIDM_DMO.pdf', bbox_inches='tight')
plt.show()
```
