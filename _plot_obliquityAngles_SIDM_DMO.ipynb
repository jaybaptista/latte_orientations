```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.family':'serif'})
```

```python
m12i_data = pd.read_hdf('_data_obliquityAngles_z0_m12i.h5')
m12f_data = pd.read_hdf('_data_obliquityAngles_z0_m12f.h5')
m12m_data = pd.read_hdf('_data_obliquityAngles_z0_m12m.h5')
m12i_data_DMO = pd.read_hdf('_data_obliquityAngles_z0_DMO_m12i.h5')
m12f_data_DMO = pd.read_hdf('_data_obliquityAngles_z0_DMO_m12f.h5')
m12m_data_DMO = pd.read_hdf('_data_obliquityAngles_z0_DMO_m12m.h5')
m12i_data_SIDM = pd.read_hdf('_data_obliquityAngles_z0_SIDM_m12i.h5')
m12f_data_SIDM = pd.read_hdf('_data_obliquityAngles_z0_SIDM_m12f.h5')
m12m_data_SIDM = pd.read_hdf('_data_obliquityAngles_z0_SIDM_m12m.h5')
```

```python
fig, ax = plt.subplots(3, dpi = 200, figsize = (8, 4.5))

r = np.arange(10, 410, step = 10)

ax_f, ax_i, ax_m = ax


ax_f.plot(r, m12f_data['angle'],
         ms=2, c='k', alpha=.5,  label='CDM+Stars')

ax_f.plot(r, m12f_data_DMO['angle'],
         ms=2, c='blue',  label='DMO')

ax_f.plot(r, m12f_data_SIDM['angle'],
         ms=2, c='dodgerblue',  label='SIDM+Stars')

ax_i.plot(r, m12i_data['angle'],
         ms=2, c='k', alpha=.5,  label='m12i')

ax_i.plot(r, m12i_data_DMO['angle'],
         ms=2, c='blue',  label='m12i')

ax_i.plot(r, m12i_data_SIDM['angle'],
         ms=2, c='dodgerblue',  label='m12i')

ax_m.plot(r, m12m_data['angle'],
         ms=2, c='k', alpha=.5,  label='m12m')

ax_m.plot(r, m12m_data_DMO['angle'],
         ms=2, c='blue',  label='m12m')

ax_m.plot(r, m12m_data_SIDM['angle'],
         ms=2, c='dodgerblue',  label='m12m')

ax_f.legend(loc = 'upper left', frameon=False)

ax_i.axhline(20, c='k', alpha=.25)
ax_m.axhline(20, c='k', alpha=.25)

for ax_k in ax:
    ax_k.axvline(60, ls='--', alpha=.25, c='k')
    ax_k.axvline(200, ls='--', alpha=.25, c='k')
    ax_k.axvline(400, ls='--', alpha=.25, c='k')


fig.tight_layout()
fig.subplots_adjust(hspace=0)
plt.show()
plt.savefig('obliquityAngles_z0_SIDM_DMO.pdf')
```
