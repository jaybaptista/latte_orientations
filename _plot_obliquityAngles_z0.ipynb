```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.family':'serif'})
```

```python
m12i_data = pd.read_hdf('_data_obliquityAngles_z0_m12i.h5')
m12f_data = pd.read_hdf('_data_obliquityAngles_z0_m12f.h5')
m12w_data = pd.read_hdf('_data_obliquityAngles_z0_m12w.h5')
m12m_data = pd.read_hdf('_data_obliquityAngles_z0_m12m.h5')
```

```python
fig, ax = plt.subplots(1, dpi = 200, figsize = (9, 3.5), sharey = 'row')
r = np.arange(10, 410, step = 10)

ax_f = ax_i = ax_w = ax_m = ax

ax_f.plot(r, m12f_data['angle'],
         ms=2, c='blue', marker=10, label='m12f')

ax_i.plot(r, m12i_data['angle'],
         ms=2, c='red', marker='o', label='m12i', ls='dashed')

ax_m.plot(r, m12m_data['angle'],
         ms=2, c='green', marker='*', label='m12m', ls='dotted')

ax_w.plot(r, m12w_data['angle'],
         ms=5, c='orange', marker='+', label='m12w', ls='dashdot')

ax_f.set_xlabel('r [kpc]')
ax_f.set_ylabel('$\\theta$ [deg]')

ax.legend(loc = 'upper left', frameon=False)

plt.show()
plt.savefig('obliquityAngles_z0.pdf')
plt.show()
```

```python
m12i_t_data = pd.read_hdf('_data_triax_z0_m12i.h5')
m12f_t_data = pd.read_hdf('_data_triax_z0_m12f.h5')
m12w_t_data = pd.read_hdf('_data_triax_z0_m12w.h5')
m12m_t_data = pd.read_hdf('_data_triax_z0_m12m.h5')
```

```python
fig, axs = plt.subplots(2, dpi = 200, figsize = (13, 3.5))
r = np.arange(10, 410, step = 10)

axs[0].plot(r, m12f_t_data['t'],
         ms=2, c='blue', marker=10, label='m12f')

axs[0].plot(r, m12i_t_data['t'],
         ms=2, c='red', marker='o', label='m12i', ls='dashed')

axs[0].plot(r, m12m_t_data['t'],
         ms=2, c='green', marker='*', label='m12m', ls='dotted')

axs[0].plot(r, m12w_t_data['t'],
         ms=5, c='orange', marker='+', label='m12w', ls='dashdot')

axs[1].plot(r, m12f_data['angle'],
         ms=2, c='blue', marker=10, label='m12f')

axs[1].plot(r, m12i_data['angle'],
         ms=2, c='red', marker='o', label='m12i', ls='dashed')

axs[1].plot(r, m12m_data['angle'],
         ms=2, c='green', marker='*', label='m12m', ls='dotted')

axs[1].plot(r, m12w_data['angle'],
         ms=5, c='orange', marker='+', label='m12w', ls='dashdot')

fig.subplots_adjust(hspace=0)
plt.show()
```
