```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io.misc.hdf5 import read_table_hdf5
import ob
```

```python
vfunc = np.vectorize(ob.cvtRedshift)
```

```python
m12i_data_vir = read_table_hdf5("Triaxialities_z/DATA_m12i_triaxiality_z_r_vir.h5")
m12f_data_vir = read_table_hdf5("Triaxialities_z/DATA_m12f_triaxiality_z_r_vir.h5")
m12w_data_vir = read_table_hdf5("Triaxialities_z/DATA_m12w_triaxiality_z_r_vir.h5")
m12m_data_vir = read_table_hdf5("Triaxialities_z/DATA_m12m_triaxiality_z_r_vir.h5")

m12i_data_rs90 = read_table_hdf5("Triaxialities_z/DATA_m12i_triaxiality_z_r_star_90.h5")
m12f_data_rs90 = read_table_hdf5("Triaxialities_z/DATA_m12f_triaxiality_z_r_star_90.h5")
m12w_data_rs90 = read_table_hdf5("Triaxialities_z/DATA_m12w_triaxiality_z_r_star_90.h5")
m12m_data_rs90 = read_table_hdf5("Triaxialities_z/DATA_m12m_triaxiality_z_r_star_90.h5")

m12i_data_5rs90 = read_table_hdf5(
    "Triaxialities_z/DATA_m12i_triaxiality_z_5r_star_90.h5"
)
m12f_data_5rs90 = read_table_hdf5(
    "Triaxialities_z/DATA_m12f_triaxiality_z_5r_star_90.h5"
)
m12w_data_5rs90 = read_table_hdf5(
    "Triaxialities_z/DATA_m12w_triaxiality_z_5r_star_90.h5"
)
m12m_data_5rs90 = read_table_hdf5(
    "Triaxialities_z/DATA_m12m_triaxiality_z_5r_star_90.h5"
)
```

```python
zs = np.arange(0, 2.1, step=0.1)
ts = vfunc(zs)
```

```python
fig, ax = plt.subplots(dpi=400)
ax.plot()
ax.plot(ts, m12f_data_vir["misc"].data[:, 6], ms=2, c="blue", marker=10, label="m12f")
ax.plot(
    ts,
    m12i_data_vir["misc"].data[:, 6],
    ms=2,
    c="red",
    marker="o",
    label="m12i",
    ls="dashed",
)
ax.plot(
    ts,
    m12m_data_vir["misc"].data[:, 6],
    ms=2,
    c="green",
    marker="*",
    label="m12m",
    ls="dotted",
)
ax.plot(
    ts,
    m12w_data_vir["misc"].data[:, 6],
    ms=5,
    c="orange",
    marker="+",
    label="m12w",
    ls="dashdot",
)
plt.show()
```

```python
fig, ax = plt.subplots(dpi=400)
ax.plot()
ax.plot(ts, m12f_data_rs90["misc"].data[:, 6], ms=2, c="blue", marker=10, label="m12f")
ax.plot(
    ts,
    m12i_data_rs90["misc"].data[:, 6],
    ms=2,
    c="red",
    marker="o",
    label="m12i",
    ls="dashed",
)
ax.plot(
    ts,
    m12m_data_rs90["misc"].data[:, 6],
    ms=2,
    c="green",
    marker="*",
    label="m12m",
    ls="dotted",
)
ax.plot(
    ts,
    m12w_data_rs90["misc"].data[:, 6],
    ms=5,
    c="orange",
    marker="+",
    label="m12w",
    ls="dashdot",
)
plt.show()
```
