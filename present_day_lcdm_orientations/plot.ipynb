```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({"font.family": "serif"})
m12i_data = pd.read_hdf("orientations_m12i_res7100.hdf")
m12f_data = pd.read_hdf("orientations_m12f_res7100.hdf")
m12w_data = pd.read_hdf("orientations_m12w_res7100.hdf")
m12m_data = pd.read_hdf("orientations_m12m_res7100.hdf")
```

```python
fig, ax = plt.subplots(1, dpi=200, figsize=(9, 3.5), sharey="row")
ax_f = ax_i = ax_w = ax_m = ax

ax_f.plot(m12f_data["radius"], m12f_data["angle"], ms=2, c="blue", label="m12f")

ax_i.plot(
    m12i_data["radius"], m12i_data["angle"], ms=2, c="red", label="m12i", ls="dashed"
)

ax_m.plot(
    m12m_data["radius"], m12m_data["angle"], ms=2, c="green", label="m12m", ls="dotted"
)

ax_w.plot(
    m12w_data["radius"],
    m12w_data["angle"],
    ms=5,
    c="orange",
    label="m12w",
    ls="dashdot",
)

ax_f.set_xlabel("r [kpc]")
ax_f.set_ylabel("$\\theta$ [deg]")

ax.legend(loc="upper left", frameon=False)

plt.show()
```
