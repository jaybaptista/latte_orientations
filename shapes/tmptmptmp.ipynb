```python
import numpy as np
import astropy.io.ascii as asc
import matplotlib.pyplot as plt
import asdf
```

```python
data_dv = asc.read("file_axis_ratios_m12f_res7100_dark.txt", format="commented_header")
data_dv_wrong = asc.read("file_axis_ratios_m12f_res7100_dark_WRONG.txt", format="commented_header")
data_dv_wronger = asc.read("file_axis_ratios_m12f_res7100_dark_EVEN_MORE_WRONG.txt", format="commented_header")

data_jb_1 = asdf.open("tmp.asdf")
data_jb_2 = asdf.open("tmp_2.asdf")
data_jb_3 = asdf.open("tmp_3.asdf")
```

```python
fig, ((ax_t, ax_s), (ax_b, ax_iter)) = plt.subplots(2, 2, dpi=200, figsize=(10,10))


ax_t.set_title("T")
ax_s.set_title("s")
ax_b.set_title("b")
ax_iter.set_title("c")

ax_t.set_xscale("log")
ax_s.set_xscale("log")
ax_b.set_xscale("log")
ax_iter.set_xscale("log")


# dv
ax_t.plot(data_dv["r"], data_dv["T"], c='r', label="1/d^2")
ax_s.plot(data_dv["r"], data_dv["s"], c='r')
ax_b.plot(data_dv["r"], data_dv["q"] , c='r')
ax_iter.plot(data_dv["r"], data_dv["s"] , c='r')

# dv wrong
ax_t.plot(data_dv_wrong["r"], data_dv_wrong["T"], c='b', label="1/d")
ax_s.plot(data_dv_wrong["r"], data_dv_wrong["s"], c='b')
ax_b.plot(data_dv_wrong["r"], data_dv_wrong["q"] , c='b')
ax_iter.plot(data_dv_wrong["r"], data_dv_wrong["s"] , c='b')

ax_t.plot(data_dv_wronger["r"], data_dv_wronger["T"], c='g', label="Not reduced")
ax_s.plot(data_dv_wronger["r"], data_dv_wronger["s"], c='g')
ax_b.plot(data_dv_wronger["r"], data_dv_wronger["q"] , c='g')
ax_iter.plot(data_dv_wronger["r"], data_dv_wronger["s"] , c='g')

ax_t.legend()

fig.tight_layout()
```
