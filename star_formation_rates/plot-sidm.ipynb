```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
plt.rcParams['font.family'] = 'serif'
```

```python
m12f = pd.read_hdf('sidm_sfr_m12f_sidm1.hdf')
m12i = pd.read_hdf('sidm_sfr_m12i_sidm1.hdf')
m12m = pd.read_hdf('sidm_sfr_m12m_sidm1.hdf')
```

```python
m12f_steady_m = m12f.time > 5.522528457992054
m12i_steady_m = m12i.time > 8.366528457990478
m12m_steady_m = m12m.time > 5.012528457992337
```

```python
fig, axs = plt.subplots(1, 3, dpi=200, figsize=(10,3))
(axf, axi, axm) = axs

for ax in axs.flatten():
    ax.set_yscale('log')
    ax.set_ylim(1e-1, 1e2)
    ax.set_xlabel("$t$ [Gyr]")
    ax.set_ylabel("SFR [$M_\odot / \mathrm{yr}$]")


axf.plot((m12f.time)[~m12f_steady_m], (m12f.sfr)[~m12f_steady_m], c='#fb8500', label='Bursty')
axi.plot((m12i.time)[~m12i_steady_m], (m12i.sfr)[~m12i_steady_m], c='#fb8500')
axm.plot((m12m.time)[~m12m_steady_m], (m12m.sfr)[~m12m_steady_m], c='#fb8500')

axf.plot((m12f.time)[m12f_steady_m], (m12f.sfr)[m12f_steady_m], c='#023047', label='Steady')
axi.plot((m12i.time)[m12i_steady_m], (m12i.sfr)[m12i_steady_m], c='#023047')
axm.plot((m12m.time)[m12m_steady_m], (m12m.sfr)[m12m_steady_m], c='#023047')

axf.set_title('m12f (SIDM)')
axi.set_title('m12i (SIDM)')
axm.set_title('m12m (SIDM)')

axf.legend(loc='lower right', frameon=False)

fig.tight_layout()
plt.savefig("sfr_sidm_t.pdf", bbox_inches="tight")
plt.show()
```
