```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
plt.rcParams['font.family'] = 'serif'
```

```python
m12f = pd.read_hdf('m12f.hdf')
m12i = pd.read_hdf('m12i.hdf')
m12m = pd.read_hdf('m12m.hdf')
m12w = pd.read_hdf('m12w.hdf')
```

```python
m12f_steady_m = m12f.time > 7.6
m12i_steady_m = m12i.time > 9.1
m12m_steady_m = m12m.time > 3.5
m12w_steady_m = m12w.time > 8.4
```

```python
fig, axs = plt.subplots(2, 2, dpi=200)
((axf, axi), (axm, axw)) = axs

for ax in axs.flatten():
    ax.set_yscale('log')
    ax.set_ylim(1e-1, 1e2)
    ax.set_xlim(-5, 5)

axf.plot((m12f.time-10.79)[~m12f_steady_m], (m12f.sfr)[~m12f_steady_m], c='#fb8500', label='Bursty')
axi.plot((m12i.time-8.06)[~m12i_steady_m], (m12i.sfr)[~m12i_steady_m], c='#fb8500')
axm.plot((m12m.time-6.33)[~m12m_steady_m], (m12m.sfr)[~m12m_steady_m], c='#fb8500')
axw.plot((m12w.time-7.97)[~m12w_steady_m], (m12w.sfr)[~m12w_steady_m], c='#fb8500')

axf.plot((m12f.time-10.79)[m12f_steady_m], (m12f.sfr)[m12f_steady_m], c='#023047', label='Steady')
axi.plot((m12i.time-8.06)[m12i_steady_m], (m12i.sfr)[m12i_steady_m], c='#023047')
axm.plot((m12m.time-6.33)[m12m_steady_m], (m12m.sfr)[m12m_steady_m], c='#023047')
axw.plot((m12w.time-7.97)[m12w_steady_m], (m12w.sfr)[m12w_steady_m], c='#023047')

axf.set_title('m12f')
axi.set_title('m12i')
axw.set_title('m12w')
axm.set_title('m12m')

fig.text(.5, 0, '$t-t_\mathrm{peri}^\mathrm{LMC}$ [Gyr]', ha='center')
fig.text(0, .5, 'SFR [$M_\odot / \mathrm{yr}$]', va='center', rotation=90)

axf.legend(loc='lower right', frameon=False)

fig.tight_layout()
plt.savefig("sfr_cdm_tperi.pdf", bbox_inches="tight")
plt.show()
```
