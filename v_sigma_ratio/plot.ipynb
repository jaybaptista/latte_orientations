```python
import matplotlib.pyplot as plt
import asdf
import numpy as np
import matplotlib as mpl
import orientations as ori

plt.rcParams['font.family'] = 'serif'

m12f = asdf.open('v_sigma_ratios_m12f_res7100.asdf')
m12i = asdf.open('v_sigma_ratios_m12i_res7100.asdf')
m12m = asdf.open('v_sigma_ratios_m12m_res7100.asdf')
m12w = asdf.open('v_sigma_ratios_m12w_res7100.asdf')

times = ori.getSnapshotData("../../../data/latte_metaldiff/m12i_res7100/", np.array(m12f["snapshot"]))[0]

m12f_steady_m = times > 7.6
m12i_steady_m = times > 9.1
m12m_steady_m = times > 3.5
m12w_steady_m = times > 8.4
```

```python
fig, axs = plt.subplots(1, 4, dpi=200, figsize=(10, 2.5), sharey=True)
(axf, axi, axm, axw) = axs

for ax in axs.flatten():
#     ax.set_yscale('log')
#     ax.set_ylim(1e-1, 1e2)
    ax.set_xlim(-5, 5)

axf.scatter((times-10.79)[~m12f_steady_m], (m12f["v.sigma.ratio"])[~m12f_steady_m], c='#fb8500')
axi.scatter((times-8.06)[~m12i_steady_m], (m12i["v.sigma.ratio"])[~m12i_steady_m], c='#fb8500')
axm.scatter((times-6.33)[~m12m_steady_m], (m12m["v.sigma.ratio"])[~m12m_steady_m], c='#fb8500')
axw.scatter((times-7.97)[~m12w_steady_m], (m12w["v.sigma.ratio"])[~m12w_steady_m], c='#fb8500', label='Bursty')

axf.scatter((times-10.79)[m12f_steady_m], (m12f["v.sigma.ratio"])[m12f_steady_m], c='#023047')
axi.scatter((times-8.06)[m12i_steady_m], (m12i["v.sigma.ratio"])[m12i_steady_m], c='#023047')
axm.scatter((times-6.33)[m12m_steady_m], (m12m["v.sigma.ratio"])[m12m_steady_m], c='#023047')
axw.scatter((times-7.97)[m12w_steady_m], (m12w["v.sigma.ratio"])[m12w_steady_m], c='#023047', label='Steady')

axf.set_title('m12f')
axi.set_title('m12i')
axw.set_title('m12w')
axm.set_title('m12m')

fig.text(.5, -0.05, '$t-t_\mathrm{peri}^\mathrm{LMC}$ [Gyr]', ha='center', fontsize=18)
fig.text(-0.05, .5, r'$\langle v_\phi \rangle / \sigma_{v, \phi}$', va='center', rotation=90, fontsize=18)

axw.legend(loc='upper left')

fig.tight_layout()
plt.savefig("vsr_cdm_tperi.png", bbox_inches="tight")
plt.show()
```
